```python
from FireSpreadModel import FireSpreadModel
import matplotlib.pyplot as plt
```

# Simple Model

```python
# fire(50, 0.8, 0.002, 0.000001, 0.25, 50)
n = 50
# probBurningArr = [0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.7]
probBurningArr = [i/10 for i in range(1,11)]
probTreeArr = [i/100 for i in range(0,101,5)]
probLightning = 0
probImmune = 0.25

forestBefore = []
forestAfter = []
for probBurning in probBurningArr:
    forestBeforeTemp = []
    forestAfterTemp = []
    for probTree in probTreeArr:
        fireModel = FireSpreadModel(n, probTree, probBurning, probLightning, probImmune, 0, 'random', 'simple', wind=False, windDir='N',windLevel=1)
        fireModel.run()
        forestBeforeTemp.append(fireModel.grids[0].flatten().tolist())
        forestAfterTemp.append(fireModel.grids[-1].flatten().tolist())
    forestAfter.append(forestAfterTemp)
    forestBefore.append(forestBeforeTemp)

area = n**2
for k in range(len(probBurningArr)):
    beforeCount = []
    afterCount = []
    beforeOverAfter = []
    for i in range(len(forestBefore[k])):
        beforeCount.append((area - forestBefore[k][i].count(0))/area)
        afterCount.append((area - forestAfter[k][i].count(0))/area)
        if beforeCount[i] == 0:
            beforeOverAfter.append(0)
        else:
            beforeOverAfter.append(afterCount[i]/beforeCount[i])

    plt.plot(beforeCount[1:], beforeOverAfter[1:],'-' , label='Prob of Burning = {}'.format(probBurningArr[k]))


plt.ylabel('Forest Density Left with respect to Intial Density')
plt.xlabel('Density of Trees')
plt.grid(True)
plt.legend()

plt.figure()

for k in range(len(probBurningArr)):
    beforeCount = []
    afterCount = []
    beforeOverAfter = []
    for i in range(len(forestBefore[k])):
        beforeCount.append((area - forestBefore[k][i].count(0))/area)
        afterCount.append((area - forestAfter[k][i].count(0))/area)
    plt.plot(beforeCount, afterCount,'-' , label='Prob of Burning = {}'.format(probBurningArr[k]))
plt.ylabel('Forest Density Left')
plt.xlabel('Density of Trees')
plt.legend()
plt.grid(True)
```
